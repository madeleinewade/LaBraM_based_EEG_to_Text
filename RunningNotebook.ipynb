```python
from math import nan
import os
import mne
import numpy as np
import pandas as pd
import pickle as pkl
import random
from huggingface_hub import snapshot_download


folder = '/root/.cache/huggingface/hub/datasets--madeleinewade--Murphy2022Data/snapshots/4783e66678539b46776b9157a20e41aed009c60a/'


x = []
y = []
OOV = []
vocabulary = {}
word_idx = 0
full_sentences = []
full_sentences_eeg = []

POS_map = {'PAD': 0, 'ADJ': 1, 'ADP': 2, 'ADV': 3, 'AUX': 4,
'CCONJ': 5, 'DET': 6, 'INTJ': 7, 'NOUN': 8, 'NUM': 9,
'PART': 10, 'PRON': 11, 'PROPN': 12, 'PUNCT': 13, 'SCONJ': 14,
'VERB': 15, 'X': 16}

print('Creating dataset...')
for f in os.listdir(folder):
    if '.fif' not in f:
        print('\'fif\' not in ', f)
        continue
    print('Loading file: ', f)
    #load epochs
    filename = folder+f
    epochs = mne.read_epochs(filename)
    epochs = epochs.resample(181)
    epochs_df = epochs.to_data_frame()
    metadata = epochs.metadata
    metadata['pos'] = metadata['pos'].astype('category')
    metadata['pos'] = metadata['pos'].map(POS_map)

    metadata['len'] = metadata['len'].astype('int') - 1

    metadata['freq'] = metadata['freq'].fillna(0)
    metadata['freq'] = metadata['freq']*10 - 20
    metadata['freq'] = metadata['freq'].astype('int')
    metadata['sent_ident'] = metadata['sent_ident'].astype(str)

    for s in np.unique(metadata['sent_ident']):
        sentence_meta = metadata[metadata['sent_ident']==s]
        sentence_eeg = epochs_df[epochs_df['epoch'].isin(sentence_meta.index)]
        sentence_x = np.zeros((32,59,200))
        sentence_y = np.zeros((32, 4))
        length = len(np.unique(sentence_eeg['epoch']))

        if length<5 or length>29:
            #skip sentences that are too long or too short
            continue

        i = 0

        full_sent = ''
        for w in np.unique(sentence_eeg['epoch']):

            meta = sentence_meta.loc[w]

            word_vec = np.empty(4)

            full_sent = full_sent + ' ' + meta['word']

            if meta['word'] not in vocabulary:
                vocabulary[meta['word']] = word_idx
                word = word_idx
                word_idx = word_idx + 1
            else:
                word = vocabulary[meta['word']]

            word_vec[0] = meta['pos']

            if meta['len'] > 20:
                word_vec[1] = 20
            else:
                word_vec[1] = meta['len']-1

            if meta['freq'] < 0:
                word_vec[2] = 0
            elif meta['freq'] > 70:
                word_vec[2] = 70
            else:
                word_vec[2] = meta['freq']
            word_vec[3] = word

            assert word_vec[1]<21
            assert word_vec[2]<71

            y.append(word_vec)

            eeg = sentence_eeg[sentence_eeg['epoch'] == w]
            eeg = eeg.drop(columns=['epoch', 'condition', 'time'])
            eeg = eeg.to_numpy()
            eeg = eeg.T

            assert eeg.shape == (64, 200)

            x.append(eeg)

            i = i + 1

        full_sentences.append(full_sent)

# print('Creating x file...')
# f= open('x.pkl', 'wb')
# pkl.dump(x, f)
# f.close()

# print('Creating y file...')
# f= open('y.pkl', 'wb')
# pkl.dump(y, f)
# f.close()

print('Creating vocabulary file...')
f= open('vocab.pkl', 'wb')
pkl.dump(vocabulary, f)
f.close()


print('Creating full sentences file...')
f= open('full_sentences.pkl', 'wb')
pkl.dump(full_sentences, f)
f.close()

print('Creating train/val/test split...')
n = len(x)
idx = list(range(n))
random.shuffle(idx)

train_idx = idx[0:int(np.round(n*0.8))]
val_idx = idx[int(np.round(n*0.8)):int(np.round(n*0.9))]
test_idx = idx[int(np.round(n*0.9)):len(idx)]

x_train = []
x_val = []
x_test = []

y_train = []
y_val = []
y_test = []

for i in range(n):
    if i in train_idx:
        x_train.append(x[i])
        y_train.append(y[i])
    elif i in test_idx:
        x_test.append(x[i])
        y_test.append(y[i])
    elif i in val_idx:
        x_val.append(x[i])
        y_val.append(y[i])
    else:
        print('error at ', i)
        print(x[i].shape)


print('Saving train/val/test split...')
f= open('/content/drive/MyDrive/LaBraM-main/x_train.pkl', 'wb')
pkl.dump(x_train, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/y_train.pkl', 'wb')
pkl.dump(y_train, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/x_val.pkl', 'wb')
pkl.dump(x_val, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/y_val.pkl', 'wb')
pkl.dump(y_val, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/x_test.pkl', 'wb')
pkl.dump(x_test, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/y_test.pkl', 'wb')
pkl.dump(y_test, f)
f.close()
```

```python
from math import nan
import os
import mne
import numpy as np
import pandas as pd
import pickle as pkl
import random

folder = '/root/.cache/huggingface/hub/datasets--madeleinewade--Murphy2022Data/snapshots/4783e66678539b46776b9157a20e41aed009c60a'

vocabulary = {'PAD':0}
word_idx = 1
full_sentences = []
sentence_x = []
y_words = []
y_length = []
y_pos = []
y_freq = []

POS_map = {'PAD': 0, 'ADJ': 1, 'ADP': 2, 'ADV': 3, 'AUX': 4,
'CCONJ': 5, 'DET': 6, 'INTJ': 7, 'NOUN': 8, 'NUM': 9,
'PART': 10, 'PRON': 11, 'PROPN': 12, 'PUNCT': 13, 'SCONJ': 14,
'VERB': 15, 'X': 16}

print('Creating dataset...')
for f in os.listdir(folder):
    if '.fif' not in f:
        print('\'fif\' not in ', f)
        continue
    print('Loading file: ', f)
    #load epochs
    filename = folder+'/'+f
    epochs = mne.read_epochs(filename)
    epochs = epochs.resample(181)
    epochs_df = epochs.to_data_frame()
    metadata = epochs.metadata
    metadata['pos'] = metadata['pos'].astype('category')
    metadata['pos'] = metadata['pos'].map(POS_map)

    metadata['len'] = metadata['len'].astype('int') - 1

    metadata['freq'] = metadata['freq'].fillna(0)
    metadata['freq'] = metadata['freq']*10 - 20
    metadata['freq'] = metadata['freq'].astype('int')
    metadata['sent_ident'] = metadata['sent_ident'].astype(str)

    for s in np.unique(metadata['sent_ident']):
        sentence_meta = metadata[metadata['sent_ident']==s]
        sentence_eeg = epochs_df[epochs_df['epoch'].isin(sentence_meta.index)]
        length = len(np.unique(sentence_eeg['epoch']))

        if length<5 or length>32:
            #skip sentences that are too long or too short
            continue

        full_sent_x = np.zeros((64,1))
        full_sent_y_words = np.array([0])
        full_sent_y_length = np.array([0])
        full_sent_y_pos = np.array([0])
        full_sent_y_freq = np.array([0])
        sentence = ''

        for w in np.unique(sentence_eeg['epoch']):

            meta = sentence_meta.loc[w]

            sentence = sentence + meta['word'] + ' '

            if meta['word'] not in vocabulary:
                vocabulary[meta['word']] = word_idx
                word = word_idx
                word_idx = word_idx + 1
            else:
                word = vocabulary[meta['word']]

            full_sent_y_words = np.append(full_sent_y_words, word)
            full_sent_y_length = np.append(full_sent_y_length, meta['len'])
            full_sent_y_pos = np.append(full_sent_y_pos, meta['pos'])
            full_sent_y_freq = np.append(full_sent_y_freq, meta['freq'])

            eeg = sentence_eeg[sentence_eeg['epoch'] == w]
            eeg = eeg.drop(columns=['epoch', 'condition', 'time'])
            eeg = eeg.to_numpy()
            eeg = eeg.T
            eeg = eeg[:, 36:80]
            full_sent_x = np.append(full_sent_x, eeg, axis = 1)

        if length < 32:
            for i in range(32-length):
                full_sent_x = np.append(full_sent_x, np.zeros((64, 44)), axis=1)
                full_sent_y_words = np.append(full_sent_y_words, 0)
                full_sent_y_length = np.append(full_sent_y_length, 0)
                full_sent_y_pos = np.append(full_sent_y_pos, 0)
                full_sent_y_freq = np.append(full_sent_y_freq, 0)


        full_sent_x = full_sent_x[:, 1:]
        full_sent_y_words = full_sent_y_words[1:]
        full_sent_y_length = full_sent_y_length[1:]
        full_sent_y_pos = full_sent_y_pos[1:]
        full_sent_y_freq = full_sent_y_freq[1:]

        assert full_sent_y_words.shape == (32, )
        assert full_sent_y_length.shape == (32, )
        assert full_sent_y_pos.shape == (32, )
        assert full_sent_y_freq.shape == (32, )
        assert full_sent_x.shape == (64, 1408)

        full_sentences.append(sentence)
        sentence_x.append(full_sent_x)
        y_words.append(full_sent_y_words)
        y_length.append(full_sent_y_length)
        y_pos.append(full_sent_y_pos)
        y_freq.append(full_sent_y_freq)

# print('Creating x file...')
# f= open('x.pkl', 'wb')
# pkl.dump(x, f)
# f.close()

# print('Creating y file...')
# f= open('y.pkl', 'wb')
# pkl.dump(y, f)
# f.close()

print('Creating vocabulary file...')
f= open('vocab.pkl', 'wb')
pkl.dump(vocabulary, f)
f.close()


print('Creating full sentences file...')
f= open('full_sentences.pkl', 'wb')
pkl.dump(full_sentences, f)
f.close()

print('Creating train/val/test split...')
n = len(sentence_x)
idx = list(range(n))
random.shuffle(idx)

train_idx = idx[0:int(np.round(n*0.8))]
val_idx = idx[int(np.round(n*0.8)):int(np.round(n*0.9))]
test_idx = idx[int(np.round(n*0.9)):len(idx)]

x_sentence_train = []
x_sentence_val = []
x_sentence_test = []

y_words_train = []
y_words_val = []
y_words_test = []

y_length_train = []
y_length_val = []
y_length_test = []

y_pos_train = []
y_pos_val = []
y_pos_test = []

y_freq_train = []
y_freq_val = []
y_freq_test = []

for i in range(n):
    if i in train_idx:
        x_sentence_train.append(sentence_x[i])
        y_words_train.append(y_words[i])
        y_length_train.append(y_length[i])
        y_pos_train.append(y_pos[i])
        y_freq_train.append(y_freq[i])
    elif i in test_idx:
        x_sentence_test.append(sentence_x[i])
        y_words_test.append(y_words[i])
        y_length_test.append(y_length[i])
        y_pos_test.append(y_pos[i])
        y_freq_test.append(y_freq[i])
    elif i in val_idx:
        x_sentence_val.append(sentence_x[i])
        y_words_val.append(y_words[i])
        y_length_val.append(y_length[i])
        y_pos_val.append(y_pos[i])
        y_freq_val.append(y_freq[i])
    else:
        print('error at ', i)
        print(x[i].shape)


print('Saving train/val/test split...')
f= open('/content/drive/MyDrive/LaBraM-main/x_train_full_sentences.pkl', 'wb')
pkl.dump(x_sentence_train, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/y_train_words.pkl', 'wb')
pkl.dump(y_words_train, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/y_train_length.pkl', 'wb')
pkl.dump(y_length_train, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/y_train_pos.pkl', 'wb')
pkl.dump(y_pos_train, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/y_train_frequency.pkl', 'wb')
pkl.dump(y_freq_train, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/x_val_full_sentences.pkl', 'wb')
pkl.dump(x_sentence_val, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/y_val_words.pkl', 'wb')
pkl.dump(y_words_val, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/y_val_length.pkl', 'wb')
pkl.dump(y_length_val, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/y_val_pos.pkl', 'wb')
pkl.dump(y_pos_val, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/y_val_frequency.pkl', 'wb')
pkl.dump(y_freq_val, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/x_test_full_sentences.pkl', 'wb')
pkl.dump(x_sentence_test, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/y_test_words.pkl', 'wb')
pkl.dump(y_words_test, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/y_test_length.pkl', 'wb')
pkl.dump(y_length_test, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/y_test_pos.pkl', 'wb')
pkl.dump(y_pos_test, f)
f.close()

f= open('/content/drive/MyDrive/LaBraM-main/y_test_frequency.pkl', 'wb')
pkl.dump(y_freq_test, f)
f.close()

```

```python
#Create custom dataset class
import torch

class EEGDataset(torch.utils.data.Dataset):
    def __init__(self, words, sent, labels):
        self.words = words
        self.sent = sent
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        word = self.words[idx]
        sent = self.sent[idx]
        label = self.labels[idx]
        return word, sent, label
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class FullEEGDecoder(nn.Module):
    def __init__(self, n_samples_per_sent, n_samples_per_word, n_channels, ch_names, vocab_size, length_model, pos_model, frequency_model):
        super(FullEEGDecoder, self).__init__()

        self.input_chans = utils.get_input_chans(ch_names)
        self.n_samples_per_sent = n_samples_per_sent
        self.n_samples_per_word = n_samples_per_word
        self.n_channels = n_channels
        self.vocab_size = vocab_size

        self.patch_embedding = nn.Conv2d(n_channels, 200, kernel_size=(1, 200), stride=(1, 200))

        self.length_model = length_model
        self.pos_model = pos_model
        self.frequency_model = frequency_model

        encoder_layer = nn.TransformerEncoderLayer(d_model=200, nhead=1, batch_first=True)
        self.encoder = nn.TransformerEncoder(encoder_layer, num_layers=1)

        decoder_layer = nn.TransformerDecoderLayer(d_model=200, nhead=1, batch_first=True)
        self.decoder = nn.TransformerDecoder(decoder_layer, num_layers = 1)

        self.fc1 = nn.Linear(1400, 512)
        self.fc2 = nn.Linear(512, vocab_size*32)


        self.activation = {}

    def get_activation(self, name):
        def hook(model, input, output):
            self.activation[name] = output.detach()
        return hook

    def forward(self, words, sentence):
        #words has dimensions (32, 64, 200)
        #sentence has dimensions (64, 1408), will have final 8 samples removed so that (64, 1400)

        self.batch_size = sentence.shape[0]
        sentence = sentence[:, :, :-8]
        sentence = sentence.type(torch.float32)
        sentence = sentence.reshape(sentence.shape[0], 64, 1, 1400)

        words = words.reshape(words.shape[0]*words.shape[1], 64, 1, 200)
        words = words.type(torch.float32)

        sentence = self.patch_embedding(sentence).flatten(2).transpose(1, 2)

        self.length_model.fc_norm.register_forward_hook(self.get_activation('fc_norm'))
        length_prediction = self.length_model(words, self.input_chans)
        length = self.activation['fc_norm']

        self.pos_model.fc_norm.register_forward_hook(self.get_activation('fc_norm'))
        pos_prediction = self.pos_model(words, self.input_chans)
        pos = self.activation['fc_norm']

        self.frequency_model.fc_norm.register_forward_hook(self.get_activation('fc_norm'))
        frequency_prediction = self.frequency_model(words, self.input_chans)
        frequency = self.activation['fc_norm']

        q = torch.cat((length, pos, frequency), dim=0)
        q = q.reshape((self.batch_size, 96, q.shape[1]))

        sentence = self.encoder(sentence)
        combined = self.decoder(sentence, q)
        combined = combined.reshape(combined.shape[0], combined.shape[1]*combined.shape[2])
        combined = self.fc1(combined)
        output = self.fc2(combined)

        return output
```

```python
def train_with_val(model, num_epochs, train_loader, val_loader, loss_function, optimizer, metric, device):
    losses = []

    for epoch in range(num_epochs):
        batch_losses = []
        metric.reset()
        for words, sentences, labels in train_loader:
            words, sentences, labels = words.to(device), sentences.to(device), labels.to(device)

            batch_size = sentences.shape[0]

            outputs = model(words, sentences)

            outputs = outputs.reshape(batch_size*32, model.vocab_size)

            # Reshape labels to have dimensions [batch_size * sequence_length]
            labels = labels.view(-1)

            labels = torch.clamp(labels, 0, model.vocab_size - 1)

            # Reshape outputs to have dimensions [batch_size * sequence_length, vocab_size]
            outputs = outputs.view(-1, model.vocab_size)

            loss = loss_function(outputs, labels.type(torch.long)) # Change data type to torch.long
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            batch_losses.append(loss.item())

            metric.update(outputs, labels.type(torch.long))  # Change data type to torch.long

        train_loss = np.mean(np.array(batch_losses))
        train_acc = metric.compute().item()
        losses.append(train_loss)

        #save parameters
        torch.save(model.state_dict(), f'model_epoch_{epoch}.pth')

        with torch.no_grad():
            for images, questions, labels in val_loader:
                images, questions, labels = images.to(device), questions.to(device), labels.to(device)
                outputs = model(images, questions)

                # Reshape labels to have dimensions [batch_size * sequence_length]
                labels = labels.view(-1)

                # Reshape outputs to have dimensions [batch_size * sequence_length, vocab_size]
                outputs = outputs.view(-1, model.vocab_size)

                metric.update(outputs, labels.type(torch.long))  # Change data type to torch.long

            val_acc = metric.compute().item()

        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {train_loss:.6f}, Train Accuracy: {train_acc*100:.4f}%, Validation Accuracy: {val_acc*100:.2f}")


    return model
```

```python
import evaluate
from torchmetrics import Accuracy

def test_model(model, test_loader, device, vocabulary):

    metric=Accuracy(task="multiclass", num_classes=len(vocabulary)).to(device)
    reverse_vocabulary = {v: k for k, v in vocabulary.items()}

    true_sentences = []
    preds_sentences = []
    with torch.no_grad():
        for words, sentences, labels in test_loader:
            words, sentences, labels = words.to(device), sentences.to(device), labels.to(device)
            outputs = model(words, sentences)
            batch_size = sentences.shape[0]

            # Reshape labels to have dimensions [batch_size * sequence_length]
            true = labels

            for sentence in labels:
                new_sentence = ''
                for word in sentence:
                    new_sentence += reverse_vocabulary[int(word)] + ' '
                true_sentences.append(new_sentence)

            # Reshape labels to have dimensions [batch_size * sequence_length]
            labels = labels.view(-1)

            # Reshape outputs to have dimensions [batch_size * sequence_length, vocab_size]
            outputs = outputs.view(-1, model.vocab_size)

            metric.update(outputs, labels.type(torch.long))  # Change data type to torch.long

            preds = outputs.reshape(batch_size, 32, -1)
            preds = torch.argmax(preds, dim=2)

            for sentence in preds:
                new_sentence = ''
                for word in sentence:
                    new_sentence += reverse_vocabulary[int(word)] + ' '
                preds_sentences.append(new_sentence)

        #Calculate accuracy
        test_acc = metric.compute().item()
        print(f"Test Accuracy: {test_acc*100:.2f}")
        #calculate BLEU and ROUGE
        # Load the BLEU and ROUGE metrics
        bleu_metric = evaluate.load("bleu")
        rouge_metric = evaluate.load("rouge")

        # BLEU expects plain text inputs
        bleu_results = bleu_metric.compute(predictions=preds_sentences, references=true_sentences)
        print(f"BLEU Score: {bleu_results['bleu'] * 100:.2f}")

        # ROUGE expects plain text inputs
        rouge_results = rouge_metric.compute(predictions=preds_sentences, references=true_sentences)
        # Access ROUGE scores (no need for indexing into the result)
        print(f"ROUGE-1 F1 Score: {rouge_results['rouge1']:.2f}")
        print(f"ROUGE-L F1 Score: {rouge_results['rougeL']:.2f}")
        return true_sentences, preds_sentences
```
